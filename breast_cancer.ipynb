```python
df = df.drop(columns = ["id","Unnamed: 32"])          ## drop two columns id and unnamed
```

```python
df
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})         ##   M = 1 (malignant), B = 0 (benign) ,0 → Benign (non-cancerous)

                                                                        # 1 → Malignant (cancerous)
   ## mapping M to 1 and B to 0
```

```python
sns.boxplot(x='diagnosis', y='radius_mean', data=df)
plt.title("Radius Mean by Diagnosis")
plt.show()
```

```python
sns.pairplot(df[['diagnosis', 'radius_mean', 'texture_mean', 'area_mean', 'smoothness_mean']], hue='diagnosis')
plt.show()
```

```python
X = df.drop(columns = ["diagnosis"])            ##  x is input other than diagnosis
y = df["diagnosis"]                             ##  y is diagnosis
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler                ##  make the all points on same level like normalization = x - mean/standard deviation

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split          ## train and test the data

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
         ## test of 20 percent data and train on 80 percent data
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier

result = RandomForestClassifier(n_estimators=100, random_state=42)        ## o 100 tree
result.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report,  accuracy_score,confusion_matrix

y_proba = result.predict_proba(X_test)[:,1]    # probability of malignant
threshold = 0.25                             # lower threshold to reduce FN
y_pred = (y_proba >= threshold).astype(int)
# y_pred = result.predict(X_test)
# 
print("Accuracy:", (accuracy_score(y_test, y_pred)))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion matrix:\n",confusion_matrix(y_test,y_pred))
```

```python
importances = result.feature_importances_
feature_names = X.columns


indices = np.argsort(importances)[::-1]        ## sort the column which are more responsible for cancer


plt.figure(figsize=(12, 8))                     ## plotting 
sns.barplot(x=importances[indices], y=feature_names[indices])
plt.title("Feature Importances from Random Forest")
plt.show()     
```
